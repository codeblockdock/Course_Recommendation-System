```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from google.colab import files
data = files.upload()
```

```python
df_courses= pd.read_csv("udemy_courses_clean.csv")
```

```python
df_courses.head()
```

```python
df_courses['course_title']
```

```python
tfidf =TfidfVectorizer(stop_words ="english")
df_courses['course_title']=df_courses['course_title'].fillna(" ")


tfidf_matrix = tfidf.fit_transform(df_courses['course_title'])
```

```python
cosine_sim=linear_kernel(tfidf_matrix,tfidf_matrix)
```

```python
indices=pd.Series(df_courses.index,index=df_courses['course_title']).drop_duplicates()
```

```python

def get_recommendation_for_new_user():
    user_input = input("Enter some topic that you're interested in: ")
    user_input_vector = tfidf.transform([user_input])
    similarity_scores = linear_kernel(user_input_vector, tfidf_matrix)
    sim_indices = similarity_scores.argsort()[0][-10:][::-1]
    print("Recommended course based on your input:")
    for idx in sim_indices:
        course_name = df_courses.iloc[idx]['course_title']
        url = df_courses.iloc[idx]['url']
        rating=df_courses.iloc[idx]['num_reviews']
        level=df_courses.iloc[idx]['level']
        print(f"course_title: {course_name}\nurl: {url}\nlevel :{level}\nnum_reviews:{rating}\n")



```

```python
get_recommendation_for_new_user()
```
